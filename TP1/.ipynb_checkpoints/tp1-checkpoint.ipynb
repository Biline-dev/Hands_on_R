# TP1 : Statistique descriptive unidimensionnelle

Soit la série statiques : 

43 29 57 45 50 29 37 59 46 31 46 24 33 38 49 31

62 60 52 38 38 26 41 52 60 49 52 41 38 26 37 59

57 41 29 33 33 43 46 57 46 33 46 49 57 57 46 43

### Calcule des Effectifs, effectifs cumulés, fréquences et fréquences cumulées

```r
ma_serie = c(43, 29, 57, 45, 50, 29, 37, 59, 46, 31, 46, 24, 33, 38, 49, 31,
              62, 60, 52, 38, 38, 26, 41, 52, 60, 49, 52, 41, 38, 26, 37, 59,
              57, 41, 29, 33, 33, 43, 46, 57, 46, 33, 46, 49, 57, 57, 46, 43)

valeurs_uniques = sort(unique(ma_serie))
effectifs = sapply(valeurs_uniques, function(x) sum(ma_serie == x))
                   
"""
function(x) sum(ma_serie == x) prend un argument, x.Pour chaque valeur unique x, 
elle vérifie combien de fois x apparaît dans ma_serie.
"""
                   
effectifs_cumules = cumsum(effectifs)
frequences = effectifs / length(ma_serie)
frequences_cumules = cumsum(frequences)

resultat = data.frame(
  Age = valeurs_uniques,
  Effectifs = effectifs,
  Effectifs_cumules = effectifs_cumules,
  Frequences = round(frequences, 4) * 100,  # rrrondir les fréquences à 4 decimales
  Frequences_cumules = round(frequences_cumules, 4) * 100
)


resultat
```

### Représentations graphiques

#### Le diagramme en bâtons

```r
# Diagramme en bâtons des effectifs
barplot(resultat$Effectifs, names.arg=resultat$Age, 
        main="Diagramme en bâtons des effectifs", 
        xlab="Age", ylab="Effectifs", col="skyblue", border="black")
```

#### Diagramme cumulatif des effectifs

```r
# Diagramme cumulatif des effectifs
plot(resultat$Age, resultat$Effectifs_cumules, type="o", col="blue", 
     main="Diagramme cumulatif des effectifs", xlab="Valeurs", ylab="Effectifs cumulés",
     pch=16)  # pch=16 ajoute des points
```

### Représentations graphiques en utilisant la librairie ggplot2

```r
# Bibliothèque pour les graphiques
library(ggplot2)

# Diagramme en bâtons
ggplot(resultat, aes(x=Age, y=Effectifs)) + 
  geom_bar(stat="identity", fill="skyblue", color="black") +
  ggtitle("Diagramme en bâtons des effectifs") +
  xlab("Valeurs") + ylab("Effectifs")


# Diagramme cumulatif
# Create the plot
ggplot(resultat, aes(x = Age, y = Effectifs_cumules)) +
  geom_step() +
  labs(title = "Diagramme cumulatif des effectifs",
       x = "Age",
       y = "Cumulative Frequency")
```

### Calcule de la médiane (quantile d’ordre 1/2), les quartiles (quantile d’ordre1/4, 3/4), la moyenne, la dispersion, la variance et l’écart–type.

```r
mediane = median(ma_serie)
quartiles = quantile(ma_serie, probs = c(0.25, 0.75))
moyenne = mean(ma_serie)
ecart_interquartile = IQR(ma_serie)
etendue = max(ma_serie) - min(ma_serie)
variance = var(ma_serie)
ecart_type = sd(ma_serie)

cat("Médiane:", mediane, "\n")
cat("Q1:", quartiles[1], "and Q2:", quartiles[2], "\n")
cat("Moyenne:", moyenne, "\n")
cat("Écart interquartile:", ecart_interquartile, "\n")
cat("Étendue:", etendue, "\n")
cat("Variance:", variance, "\n")
cat("Écart-type:", ecart_type, "\n")
```
